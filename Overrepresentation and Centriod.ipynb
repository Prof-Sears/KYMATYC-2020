# A Jupyter Notebook for measuring unfairness of the Electoral College

```python
# Import Libraries
import numpy as np
import pandas as pd
```

## Load the data

Here, we load the data into a Pandas DataFrame.

```python
# Import Population and Electoral College votes
population_df = pd.read_csv('Census_Data_By_Year - 2010.csv')
votes_df = pd.read_csv('max_2010_apportionment.csv')
print(population_df)
print(votes_df)
```

## Define a function to process the data

The overrepresentation of each state is calculated. Then, the moment, votes times overrepresentation is calculated.

```python
def process_apportionment(population_df, votes_df):
    power_list = []
    
    for index, row in population_df.iterrows():
        power_list.append(votes_df.at[index,'Electoral Votes'] / row['Population'])
        #print('State: {0} Population: {1} Votes: {2} Power: {3}'.format(row['State'],row['Population'],votes_df.at[index,'Electoral Votes'], power_list[-1]))
        
    minindex = power_list.index(min(power_list))
    print('Least powerful state {0}'.format(population_df.at[minindex,'State']))
    
    data_list = []
    
    for index, row in population_df.iterrows():
        data_list.append({'State':row['State'], 'Overrepresentation': (power_list[index] - power_list[minindex])/power_list[minindex], 'Moment': votes_df.at[index,'Electoral Votes']*(power_list[index] - power_list[minindex])/power_list[minindex]})
        
    return pd.DataFrame(data_list)    
```

```python
overrep_df = process_apportionment(population_df, votes_df)
print(overrep_df)
```

```python
print('Centroid: {0}'.format(overrep_df['Moment'].sum()/votes_df['Electoral Votes'].sum()))
print('Total Electoral votes: {0}'.format(votes_df['Electoral Votes'].sum()))
```
