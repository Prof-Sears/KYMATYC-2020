```python
import math
import numpy as np
import pandas as pd
```

```python
# This is a function to do Huntington-Hill apportionment using Proportional parts
# states_df: a Pandas dataframe containing the following series: 'State', 'Population'
def Huntington_Hill(states_df, seats):
    priority_values = []
    apportionment = []
    
    for i in range(states_df['State'].count()):
        apportionment.append(1)
        priority_values.append(states_df.at[i,'Population']/math.sqrt(2))
        
    for i in range(seats):
        maxpos = priority_values.index(max(priority_values))
        print("Next seat to: {0} \t\t Priority Value: {1}".format(states_df.at[maxpos,'State'],priority_values[maxpos]))
        apportionment[maxpos] += 1
        priority_values[maxpos] = states_df.at[maxpos,'Population']/math.sqrt(apportionment[maxpos]*(apportionment[maxpos]+1))
        #print("New apportionment: {0} \t\t New Priority Value: {1}".format(apportionment[maxpos],priority_values[maxpos]))
        
    apportionment_dict = {'State':states_df['State'].values, 'Apportionment':apportionment}
    apportionment_df = pd.DataFrame(apportionment_dict)
    
    return apportionment_df
```

```python
# Read in the states populations
states_df = pd.read_csv("2000_populations.csv")
print(states_df)
```

```python
apportionment_df = Huntington_Hill(states_df, 385)
```

```python
# This is a function to do Huntington-Hill apportionment using Proportional parts
# states_df: a Pandas dataframe containing the following series: 'State', 'Population'
def MAX_Huntington_Hill(states_df):
    priority_values = []
    apportionment = []
    
    for i in range(states_df['State'].count()):
        apportionment.append(1)
        priority_values.append(states_df.at[i,'Population']/math.sqrt(2))
        
    while(min(apportionment) < 2):
        maxpos = priority_values.index(max(priority_values))
        print("Next seat to: {0} \t\t Priority Value: {1}".format(states_df.at[maxpos,'State'],priority_values[maxpos]))
        apportionment[maxpos] += 1
        priority_values[maxpos] = states_df.at[maxpos,'Population']/math.sqrt(apportionment[maxpos]*(apportionment[maxpos]+1))
        #print("New apportionment: {0} \t\t New Priority Value: {1}".format(apportionment[maxpos],priority_values[maxpos]))
        
    apportionment_dict = {'State':states_df['State'].values, 'Apportionment':apportionment}
    apportionment_df = pd.DataFrame(apportionment_dict)
    
    return apportionment_df
```

```python
max_apportionment = MAX_Huntington_Hill(states_df)
```

```python
print(max_apportionment)
print('Total Apportionment: {0}',format(max_apportionment['Apportionment'].sum()))
```
